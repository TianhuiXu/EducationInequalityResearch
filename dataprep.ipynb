```python
import pandas as pd
import scipy.stats as sts

# create the dictionary to store the data
data = {'year':[], 'schoolid':[], 'treatment':[], 'income':[],'outcome':[],'parents':[],'post':[]}

# iterate over all periods
for year in [2017, 2018, 2019, 2020, 2021, 2022]:
    # iterate over the two treatment status
    for treatment in range(0, 2):
        # generate data from distributions
        income_dst = sts.norm(treatment * 100000 + 50000, 50)
        outcome_dst = sts.norm(treatment * 30 + 600, 50)
        parents_dst = sts.norm(treatment * 3 + 9, 3)
        # append data to the dictionary
        for schoolid in range(1, 10):
            data['year'].append(year)
            data['schoolid'].append(schoolid + 10 * treatment)
            data['treatment'].append(treatment)
            data['income'].append(income_dst.rvs())
            data['outcome'].append(outcome_dst.rvs())
            data['parents'].append(parents_dst.rvs())
            if year <= 2019:
                data['post'].append(1)
            else:
                data['post'].append(0)

# transform into dataframe
df = pd.DataFrame(data)
df.head()
```

```python
df.to_stata('test.dta')  
```
